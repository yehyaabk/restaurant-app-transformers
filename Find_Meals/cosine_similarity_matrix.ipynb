This notebook explores the Indian Food 101 dataset.

## Objectives

- Create text embeddings for each product (dish).
- Store these embeddings for later use.

These embeddings allow recommendations based on meaning, not just exact word matching.

## Cosine Similarity Matrix

We also compute a cosine similarity matrix between all products.

- Each row represents one product.
- Each column represents another product.
- Each value shows how similar two products are in meaning.
- Each value `sim[i][j]` indicates how **semantically similar** product *i* is to product *j*

This matrix will be used to recommend dishes, especially based on previous user orders.
